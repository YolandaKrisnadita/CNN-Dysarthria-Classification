```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
import pickle
import keras
import keras.utils
from tqdm import tqdm
import IPython.display as ipd
```

# Load The Model

```python
#load model ML
with open('rf_model.pkl', 'rb') as file_1:
  rf_model = pickle.load(file_1)
# scaler = StandardScaler()  
with open('model_scaler.pkl', 'rb') as file_2:
  scaler = pickle.load(file_2)

# Load the SCC model
cnn_model = tf.keras.models.load_model('my_saved_model.h5')
```

# Make Function for prediction

```python
# Define function for inference using Random Forest model
def rf_inference(audio_file):
    # Load audio file
    audio, sr = librosa.load(audio_file, sr=None)
    # Perform feature Extraction.
    mfcc = np.mean(librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=128),axis=1)
    # # Flatten MFCC features
    mfcc_flat = mfcc.reshape(1, -1)
    #Normalize features using the scaler
    mfcc_scaled = scaler.transform(mfcc_flat)
    # Reshape features back to original shape
    #features = process_audio(mfcc_flat)
    # Perform inference
    prediction = rf_model.predict(mfcc_scaled)
    return prediction

# Define function for inference using CNN model
def cnn_inference(audio_file):
    # Load audio file
    audio, sr = librosa.load(audio_file, sr=None)
    # Perform feature Extraction.
    mfcc = np.mean(librosa.feature.mfcc(y=audio, sr=sr,n_mfcc=128),axis=1)
    # Reshape features for CNN input
    features = mfcc.reshape(-1,16,8,1)
    # Perform inference
    prediction = cnn_model.predict(features)
    prediction[prediction>=0.5] = 1
    prediction[prediction<0.5] = 0
    return prediction
```
